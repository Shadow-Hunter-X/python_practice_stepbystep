```python
from os import path
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
hadoop_mask = np.array(Image.open(path.join('./', "66.png")))
text = "hadoop spark python"
```

```python
wc = WordCloud(background_color="white", repeat=True, mask=hadoop_mask)
wc.generate(text)

plt.axis("off")
plt.imshow(wc, interpolation="bilinear")
plt.show()
```

```python
wc.to_file("first.png")
```
