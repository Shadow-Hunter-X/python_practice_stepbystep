```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = np.arange(1, 7)
plt.plot(y-1)
plt.plot(y)
plt.plot(y+1)
plt.plot(y-1,y,y+1)
```

```python
plt.plot(y, 'r', y+1, 'y', y+2, 'b') 
```

```python
plt.plot(y, ':', y+1, '--', y+2, '-.')
```

```python
plt.plot(y, '.', y+1, 'v', y+2, 'p')
```

```python
plt.plot(y, 'r:.', y+1, 'y--v', y+2, 'b-.p')
```

```python
plt.plot(y-1, color='r', linestyle=':', linewidth=3,
marker='.', markerfacecolor='red', markeredgecolor='y',
markeredgewidth=3, markersize=5,label='y-1')
plt.plot(y, color='y', linestyle='--', linewidth=3,
marker='v', markerfacecolor='y', markeredgecolor='b',
markeredgewidth=3, markersize=5,label='y-1')
plt.plot(y+1, color='b', linestyle='-.', linewidth=3,
marker='p', markerfacecolor='b', markeredgecolor='r',
markeredgewidth=3, markersize=5,label='y-1')
plt.grid(True)  #显示图片网格
plt.axis()      #显示当前轴限制值
plt.xlabel(u'X Value')
plt.ylabel(u'Y Value')   #设置XY轴的说明
plt.title('plot test')   #添加图片标题栏
plt.legend(['y-1','y','y+1'])   #添加说明，也可以在plot函数中，使用label参数
plt.xticks(range(len(y)), ['a', 'b', 'c', 'd', 'e', 'f','g'])
plt.yticks(range(1, 8, 2))      #设置X、Y轴的刻度
plt.savefig('plot_test.png')    #保存图片
```
